```python
import numpy as np

from ryd_gate.lattice import make_square_lattice, plot_spatial_rydberg


geom = make_square_lattice(4, 4, spacing_um=4.0)
rydberg_occ = np.where(geom.sublattice > 0, 1.0, 0.0)

fig = plot_spatial_rydberg(
    geom.coords,
    rydberg_occ,
    sublattice=geom.sublattice,
    title="4x4 square lattice",
)
```
